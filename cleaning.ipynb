```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("oleksiimartusiuk/bts-january-2024-commercial-flights-data")

print("Path to dataset files:", path)
```

```python
df = pd.read_csv(path + "/T_ONTIME_MARKETING.csv")
df.head(10)
```

```python
df.columns.tolist()
```

```python
df.shape
```

```python
null_percentages = (df.isnull().sum() / len(df)) * 100
validate_summary = pd.DataFrame({
    'data type': df.dtypes,
    'null count': df.isnull().sum(),
    'null %': null_percentages
})

validate_summary = validate_summary.sort_values(by='null %', ascending=False)
validate_summary.head(59)
```

```python
mask = (df['CANCELLED'] != 1) & (df['DIVERTED'] != 1)
df = df[mask]
```

```python
columns_to_drop = null_percentages[null_percentages > 10].index
df = df.drop(columns=columns_to_drop)
df.columns.tolist()
```

```python
df = df.drop(columns=['TAIL_NUM', 'OP_UNIQUE_CARRIER', 'ORIGIN_CITY_MARKET_ID', 'DEST_CITY_MARKET_ID', 'TAXI_OUT','TAXI_IN','WHEELS_OFF', 'WHEELS_ON','DUP', 'DIVERTED', 'CANCELLED', 'DIV_AIRPORT_LANDINGS', 'OP_CARRIER_FL_NUM'])
df.columns.tolist()
```

```python
df = df.dropna()
df.shape
```

```python
df.to_csv('jan24flights_cleaned.csv', index=False)
```
